# arithmetic in numpy

```python
import numpy as np
```

```python
# scalar arithmetic
arr=np.array([1,2,3,4])

# we can (* + - % /) to the array with only one line like the following: you can replace the * with any one you want 
# also you can make the powe like this arr ** 2
arr * 3
```

```python
arr2=np.array([1.2,2.5,3.99])
# numpy also have vectorized math functions
np.sqrt(arr2)
np.round(arr2) #-> and so on on the math function you have in python you can use it like this 
np.floor(arr2)

np.pi #-> pi value
```

```python
# quick exercise 
radius = np.array([1,2,3])

# pi * r**2 -> apply thid equation to all the array
np.pi * (radius**2)
```

# element wise arithmetic

```python
# with element wise arithmetic we can apply operationc between 2 different arrays 

arr4=np.array([1,2,3])
arr5=np.array([4,5,6])
arr4+arr5
arr4*arr5
```

# comparison operation

```python
# with comparison operation we can make filtertion to the data or get boolean values from it and so on 
arr6=np.array([1,2,3,4,5,6,7,78,8,9,9,0])
arr6[arr6 > 3]  
# -> look inside the bracket we add the comparison operation we want to make and then add 
# it in the loop bracket to get the values because what will return without the array bracket is a boolean values
```
