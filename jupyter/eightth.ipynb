# Random function in numpy  : 
### we use it to generate a random integers ..etc to make the operations we want

```python
import numpy as np
```

```python
rng=np.random.default_rng()
```

```python
# this will create a random number from 1 to 6 because the last number will not count
print(rng.integers(low=1,high=7,size=3)) #-> size here is responsible of the number of values that will be returned : (low-high) is not necessary
```

```python
# what if we want to make a 2d array 
print(rng.integers(low=1,high=11,size=(3,2)))
```

```python
# seed in default_rng make you save the random values that have generated to use it 
# it is like an id to your array
rng1 = np.random.default_rng(seed=3)
print(rng1.integers(0, 10, 5))

rng2 = np.random.default_rng(seed=3)
print(rng2.integers(0, 10, 5))
```

# Uniform function

```python
# in numpy uniform function is used to generate randoom numbers (uniformally distributed)
rng4=rng.uniform(10,20)
print(rng4) # -> will return a random value in the range of 10 and 20
```

```python
# the size in uniform 
arr=rng.uniform(0,1,5)
print(arr) #-> will return an array of 5 numbers from the range of 0 to 1
```

# choice : to make a random choice and print it

```python
rng6=np.random.default_rng()
fruits=np.array(["apple","banana","kiwi","coconut","pineapple","lemon"])

fruit_choice=rng6.choice(fruits)
print(fruit_choice)
```

```python
# to make an array of a random choices  
rng7=np.random.default_rng()
fruits=np.array(["apple","banana","kiwi","coconut","pineapple","lemon"])

fruit_choice=rng6.choice(fruits,size=3)

print(fruit_choice)
```
