# Multidimensional arrays in numpy

## ndim -> return the number of the dimensions
## shape -> return a tuple of layers-rows-columns


```python
arr=np.array('s')
arr1=np.array([1,2])

arr2=np.array([[1,2,3],
                [1,2,3]])

arr3=np.array([[[1,2,3],
                [1,2,3]]])

print(arr.ndim)
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

print(arr.shape)
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

# if we want to print specific element you can use chain indexing but it is slow
print(arr3[0][0][1]) #-> this will return 2
print(arr2[0][2])
```

```python
# welcome to multidemnsional indexing 
print(arr3[0,0,1]) # will also return 2 but it is more way faster than the chain indexing
```
